Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
!unzip -q /content/drive/MyDrive/Dataset.zip -d /content/
```

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# show first 7 rows of the dataset
credit_card_data.head(7)
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Division'].value_counts()
```

0 --> Normal Transaction

1 --> fraudulent transaction

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Division == 0]
fraud = credit_card_data[credit_card_data.Division == 1]
```

```python
# statistical measures of the data
legit.Amt.describe()
```

```python
fraud.Amt.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Division').mean()
```

Under-Sampling

```python
legit_sample = legit.sample(n=492)
```

Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head(7)
```

```python
new_dataset.tail(7)
```

```python
new_dataset['Division'].value_counts()
```

```python
new_dataset.groupby('Division').mean()
```

Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Division', axis=1)
Y = new_dataset['Division']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

```python
recall_score(X_test_prediction,Y_test)
```

```python
precision_score(X_test_prediction,Y_test)
```

```python
f1_score(X_test_prediction,Y_test)
```
